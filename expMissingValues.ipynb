```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
allData = [
    pd.read_csv("../data-gml-health/data-gml-health/1021/asthma_incidence_rate.csv"),
    pd.read_csv("../data-gml-health/data-gml-health/1021/asthma_prevalence_rate.csv"),
    pd.read_csv("../data-gml-health/data-gml-health/1021/COPD_incidence_rate.csv"),
    pd.read_csv("../data-gml-health/data-gml-health/1021/COPD_prevalence_rate.csv"),
    pd.read_csv("../data-gml-health/data-gml-health/1021/tuberculosis_incidence_rate.csv"),
    pd.read_csv("../data-gml-health/data-gml-health/1021/tuberculosis_prevalence_rate.csv"),
]

concatDf = pd.concat(allData)

droppedConcat = concatDf.drop(columns=["Measure", "Disease", "Metric", "Value"])

droppedConcat
```

```python
droppedConcat = droppedConcat.drop_duplicates()

navalues = droppedConcat.isna().sum().sort_values()

navalues = navalues[navalues > 0]

plt.figure()
plt.barh(navalues.index, navalues.values)
```
