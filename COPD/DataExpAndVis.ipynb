## 4. Analyse des valeurs manquantes

```python
# Analyse des valeurs manquantes
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Colonne': missing.index,
    'Valeurs manquantes': missing.values,
    'Pourcentage': missing_pct.values
})

# Filtrer uniquement les colonnes avec des valeurs manquantes
missing_df = missing_df[missing_df['Valeurs manquantes'] > 0].sort_values('Pourcentage', ascending=False)

print("‚ö†Ô∏è Valeurs manquantes par colonne:")
print(f"{'='*70}")
if len(missing_df) > 0:
    print(missing_df.to_string(index=False))
    print(f"\nüìä Total de valeurs manquantes: {df.isnull().sum().sum()}")
else:
    print("‚úÖ Aucune valeur manquante d√©tect√©e!")
```

```python
# Visualisation des valeurs manquantes
if len(missing_df) > 0:
    top_missing = missing_df.head(15)
    
    plt.figure(figsize=(14, 6))
    plt.barh(top_missing['Colonne'], top_missing['Pourcentage'], color='coral')
    plt.xlabel('Pourcentage de valeurs manquantes (%)', fontsize=12)
    plt.title('Top 15 colonnes avec valeurs manquantes', fontsize=14, fontweight='bold')
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("‚úÖ Pas de visualisation n√©cessaire (aucune valeur manquante)")
```

## 5. Distribution de la variable cible (Value)

```python
# Statistiques sur la variable cible
print("üìä Statistiques de la variable 'Value':")
print(f"{'='*70}")
print(f"Moyenne: {df['Value'].mean():.4f}")
print(f"M√©diane: {df['Value'].median():.4f}")
print(f"√âcart-type: {df['Value'].std():.4f}")
print(f"Min: {df['Value'].min():.4f}")
print(f"Max: {df['Value'].max():.4f}")
print(f"Quartile 25%: {df['Value'].quantile(0.25):.4f}")
print(f"Quartile 75%: {df['Value'].quantile(0.75):.4f}")

# Distribution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogramme
axes[0].hist(df['Value'].dropna(), bins=50, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].axvline(df['Value'].mean(), color='red', linestyle='--', linewidth=2, label=f'Moyenne: {df["Value"].mean():.2f}')
axes[0].axvline(df['Value'].median(), color='green', linestyle='--', linewidth=2, label=f'M√©diane: {df["Value"].median():.2f}')
axes[0].set_xlabel('Value', fontsize=11)
axes[0].set_ylabel('Fr√©quence', fontsize=11)
axes[0].set_title('Distribution de Value (Histogramme)', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Box plot
axes[1].boxplot(df['Value'].dropna(), vert=True)
axes[1].set_ylabel('Value', fontsize=11)
axes[1].set_title('Distribution de Value (Box Plot)', fontsize=12, fontweight='bold')
axes[1].grid(alpha=0.3)

# Violin plot
axes[2].violinplot([df['Value'].dropna()], vert=True, showmeans=True, showmedians=True)
axes[2].set_ylabel('Value', fontsize=11)
axes[2].set_title('Distribution de Value (Violin Plot)', fontsize=12, fontweight='bold')
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. √âvolution temporelle

```python
# √âvolution temporelle globale (moyenne par ann√©e)
yearly_stats = df.groupby('Year')['Value'].agg(['mean', 'median', 'std', 'count']).reset_index()

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Graphique 1: Moyenne et m√©diane
axes[0].plot(yearly_stats['Year'], yearly_stats['mean'], marker='o', linewidth=2.5, 
             markersize=7, label='Moyenne', color='steelblue')
axes[0].plot(yearly_stats['Year'], yearly_stats['median'], marker='s', linewidth=2.5, 
             markersize=7, label='M√©diane', color='coral')
axes[0].fill_between(yearly_stats['Year'], 
                      yearly_stats['mean'] - yearly_stats['std'],
                      yearly_stats['mean'] + yearly_stats['std'],
                      alpha=0.2, color='steelblue', label='¬±1 √©cart-type')
axes[0].set_xlabel('Ann√©e', fontsize=12)
axes[0].set_ylabel('Value', fontsize=12)
axes[0].set_title('√âvolution temporelle globale de COPD (Moyenne ¬± √âcart-type)', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# Graphique 2: Nombre d'observations
axes[1].bar(yearly_stats['Year'], yearly_stats['count'], color='lightgreen', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Ann√©e', fontsize=12)
axes[1].set_ylabel('Nombre d\'observations', fontsize=12)
axes[1].set_title('Nombre d\'observations par ann√©e', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Analyse de la tendance
trend_change = yearly_stats['mean'].iloc[-1] - yearly_stats['mean'].iloc[0]
trend_pct = (trend_change / yearly_stats['mean'].iloc[0]) * 100

print(f"üìà Analyse de la tendance:")
print(f"{'='*70}")
print(f"Valeur initiale ({yearly_stats['Year'].iloc[0]}): {yearly_stats['mean'].iloc[0]:.4f}")
print(f"Valeur finale ({yearly_stats['Year'].iloc[-1]}): {yearly_stats['mean'].iloc[-1]:.4f}")
print(f"Changement: {trend_change:.4f} ({trend_pct:+.2f}%)")
if trend_change > 0:
    print("‚û°Ô∏è Tendance: AUGMENTATION")
else:
    print("‚û°Ô∏è Tendance: DIMINUTION")
```

## 7. Analyse g√©ographique (Top/Bottom pays)

```python
# Calculer la moyenne par pays
country_avg = df.groupby('Country Name')['Value'].mean().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Top 15 pays (valeurs les plus √©lev√©es)
top_15 = country_avg.head(15)
axes[0].barh(range(len(top_15)), top_15.values, color='darkred', edgecolor='black')
axes[0].set_yticks(range(len(top_15)))
axes[0].set_yticklabels(top_15.index, fontsize=10)
axes[0].set_xlabel('Valeur moyenne', fontsize=12)
axes[0].set_title('Top 15 pays - Valeurs les plus √©lev√©es', fontsize=13, fontweight='bold')
axes[0].invert_yaxis()
axes[0].grid(axis='x', alpha=0.3)

# Bottom 15 pays (valeurs les plus faibles)
bottom_15 = country_avg.tail(15).sort_values(ascending=True)
axes[1].barh(range(len(bottom_15)), bottom_15.values, color='darkgreen', edgecolor='black')
axes[1].set_yticks(range(len(bottom_15)))
axes[1].set_yticklabels(bottom_15.index, fontsize=10)
axes[1].set_xlabel('Valeur moyenne', fontsize=12)
axes[1].set_title('Bottom 15 pays - Valeurs les plus faibles', fontsize=13, fontweight='bold')
axes[1].invert_yaxis()
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üåç R√©sum√© g√©ographique:")
print(f"{'='*70}")
print(f"Pays avec valeur la plus √©lev√©e: {country_avg.index[0]} ({country_avg.iloc[0]:.4f})")
print(f"Pays avec valeur la plus faible: {country_avg.index[-1]} ({country_avg.iloc[-1]:.4f})")
print(f"Ratio max/min: {country_avg.iloc[0] / country_avg.iloc[-1]:.2f}x")
```

## 9. Analyse des corr√©lations (Features num√©riques)

```python
# Identifier les colonnes num√©riques
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Retirer les colonnes d'identification
id_cols = ['Year']
feature_cols = [col for col in numeric_cols if col not in id_cols]

print(f"üìä Colonnes num√©riques √† analyser: {len(feature_cols)}")

# Calculer les corr√©lations avec Value
if 'Value' in feature_cols:
    correlations = df[feature_cols].corr()['Value'].drop('Value').abs().sort_values(ascending=False)
    
    # Top 15 corr√©lations
    top_15_corr = correlations.head(15)
    
    print(f"\nüîó Top 15 features corr√©l√©es avec 'Value':")
    print(f"{'='*70}")
    for feat, corr in top_15_corr.items():
        print(f"{feat[:50]:50s} : {corr:.4f}")
    
    # Visualisation
    plt.figure(figsize=(12, 8))
    plt.barh(range(len(top_15_corr)), top_15_corr.values, color='teal', edgecolor='black')
    plt.yticks(range(len(top_15_corr)), [f[:40] for f in top_15_corr.index], fontsize=10)
    plt.xlabel('Corr√©lation absolue avec Value', fontsize=12)
    plt.title('Top 15 Features - Corr√©lation avec Value', fontsize=14, fontweight='bold')
    plt.grid(axis='x', alpha=0.3)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Colonne 'Value' non trouv√©e dans les donn√©es num√©riques")
```

```python
# Matrice de corr√©lation pour top features
if 'Value' in feature_cols and len(top_15_corr) > 0:
    # S√©lectionner top features + Value
    selected_features = list(top_15_corr.head(10).index) + ['Value']
    
    # Calculer matrice de corr√©lation
    corr_matrix = df[selected_features].corr()
    
    # Heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
                square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
                xticklabels=[f[:30] for f in corr_matrix.columns],
                yticklabels=[f[:30] for f in corr_matrix.index])
    plt.title('Matrice de corr√©lation - Top 10 Features vs Value', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=9)
    plt.yticks(fontsize=9)
    plt.tight_layout()
    plt.show()
```

### 9.1 Analyse des corr√©lations (Mortality Rate < 15)

```python
# Filtrer les donn√©es avec mortality rate < 15
mortality_col = 'Mortality rate, infant (per 1,000 live births)'

if mortality_col in df.columns:
    # V√©rifier si la colonne contient des valeurs num√©riques
    df_mortality_filtered = df[df[mortality_col] < 15].copy()
    
    print(f"üìä Donn√©es filtr√©es (Mortality Rate < 15):")
    print(f"{'='*70}")
    print(f"Lignes avant filtrage: {len(df)}")
    print(f"Lignes apr√®s filtrage: {len(df_mortality_filtered)}")
    print(f"Pourcentage conserv√©: {(len(df_mortality_filtered) / len(df)) * 100:.2f}%")
    print(f"Pays uniques (avant): {df['Country Name'].nunique()}")
    print(f"Pays uniques (apr√®s): {df_mortality_filtered['Country Name'].nunique()}")
    
    # Recalculer les corr√©lations pour le sous-ensemble filtr√©
    numeric_cols_filtered = df_mortality_filtered.select_dtypes(include=[np.number]).columns.tolist()
    id_cols_filtered = ['Year']
    feature_cols_filtered = [col for col in numeric_cols_filtered if col not in id_cols_filtered]
    
    if 'Value' in feature_cols_filtered:
        correlations_filtered = df_mortality_filtered[feature_cols_filtered].corr()['Value'].drop('Value').abs().sort_values(ascending=False)
        
        # Top 15 corr√©lations
        top_15_corr_filtered = correlations_filtered.head(15)
        
        print(f"\nüîó Top 15 features corr√©l√©es avec 'Value' (Mortality Rate < 15):")
        print(f"{'='*70}")
        for feat, corr in top_15_corr_filtered.items():
            print(f"{feat[:50]:50s} : {corr:.4f}")
        
        # Visualisation compar√©e
        fig, axes = plt.subplots(1, 2, figsize=(18, 8))
        
        # Toutes les donn√©es
        axes[0].barh(range(len(top_15_corr)), top_15_corr.values, color='steelblue', edgecolor='black', alpha=0.7)
        axes[0].set_yticks(range(len(top_15_corr)))
        axes[0].set_yticklabels([f[:35] for f in top_15_corr.index], fontsize=9)
        axes[0].set_xlabel('Corr√©lation absolue avec Value', fontsize=12)
        axes[0].set_title('Top 15 Features - Toutes les donn√©es', fontsize=13, fontweight='bold')
        axes[0].grid(axis='x', alpha=0.3)
        axes[0].invert_yaxis()
        
        # Donn√©es filtr√©es
        axes[1].barh(range(len(top_15_corr_filtered)), top_15_corr_filtered.values, color='coral', edgecolor='black', alpha=0.7)
        axes[1].set_yticks(range(len(top_15_corr_filtered)))
        axes[1].set_yticklabels([f[:35] for f in top_15_corr_filtered.index], fontsize=9)
        axes[1].set_xlabel('Corr√©lation absolue avec Value', fontsize=12)
        axes[1].set_title('Top 15 Features - Mortality Rate < 15', fontsize=13, fontweight='bold')
        axes[1].grid(axis='x', alpha=0.3)
        axes[1].invert_yaxis()
        
        plt.tight_layout()
        plt.show()
        
        # Matrice de corr√©lation pour top features (donn√©es filtr√©es)
        selected_features_filtered = list(top_15_corr_filtered.head(10).index) + ['Value']
        corr_matrix_filtered = df_mortality_filtered[selected_features_filtered].corr()
        
        # Heatmap
        plt.figure(figsize=(12, 10))
        sns.heatmap(corr_matrix_filtered, annot=True, fmt='.2f', cmap='coolwarm', center=0,
                    square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
                    xticklabels=[f[:30] for f in corr_matrix_filtered.columns],
                    yticklabels=[f[:30] for f in corr_matrix_filtered.index])
        plt.title('Matrice de corr√©lation - Mortality Rate < 15', fontsize=14, fontweight='bold')
        plt.xticks(rotation=45, ha='right', fontsize=9)
        plt.yticks(fontsize=9)
        plt.tight_layout()
        plt.show()
    else:
        print("‚ö†Ô∏è Colonne 'Value' non trouv√©e dans les donn√©es filtr√©es")
else:
    print(f"‚ö†Ô∏è Colonne '{mortality_col}' non trouv√©e dans le dataset")
    print(f"\nColonnes disponibles contenant 'Mortality':")
    mortality_cols = [col for col in df.columns if 'mortality' in col.lower()]
    for col in mortality_cols:
        print(f"  - {col}")
```

### 9.2 Analyse des corr√©lations (Mortality Rate < 15 ET GDP < 0.5)

```python
# Filtrer les donn√©es avec mortality rate < 15 ET GDP < 0.5
gdp_col = 'GDP (current US$)'

if mortality_col in df.columns and gdp_col in df.columns:
    # Appliquer les deux filtres
    df_mortality_gdp_filtered = df[(df[mortality_col] < 15) & (df[gdp_col] < 0.5e13)].copy()
    
    print(f"üìä Donn√©es filtr√©es (Mortality Rate < 15 ET GDP < 0.5):")
    print(f"{'='*70}")
    print(f"Lignes avant filtrage: {len(df)}")
    print(f"Lignes apr√®s filtrage: {len(df_mortality_gdp_filtered)}")
    print(f"Pourcentage conserv√©: {(len(df_mortality_gdp_filtered) / len(df)) * 100:.2f}%")
    print(f"Pays uniques (avant): {df['Country Name'].nunique()}")
    print(f"Pays uniques (apr√®s): {df_mortality_gdp_filtered['Country Name'].nunique()}")
    
    if len(df_mortality_gdp_filtered) > 0:
        # Recalculer les corr√©lations pour ce sous-ensemble
        numeric_cols_gdp = df_mortality_gdp_filtered.select_dtypes(include=[np.number]).columns.tolist()
        id_cols_gdp = ['Year']
        feature_cols_gdp = [col for col in numeric_cols_gdp if col not in id_cols_gdp]
        
        if 'Value' in feature_cols_gdp:
            correlations_gdp = df_mortality_gdp_filtered[feature_cols_gdp].corr()['Value'].drop('Value').abs().sort_values(ascending=False)
            
            # Top 15 corr√©lations
            top_15_corr_gdp = correlations_gdp.head(15)
            
            print(f"\nüîó Top 15 features corr√©l√©es avec 'Value' (Mortality < 15 & GDP < 0.5):")
            print(f"{'='*70}")
            for feat, corr in top_15_corr_gdp.items():
                print(f"{feat[:50]:50s} : {corr:.4f}")
            
            # Visualisation compar√©e (3 graphiques)
            fig, axes = plt.subplots(1, 3, figsize=(20, 8))
            
            # Toutes les donn√©es
            axes[0].barh(range(len(top_15_corr)), top_15_corr.values, color='steelblue', edgecolor='black', alpha=0.7)
            axes[0].set_yticks(range(len(top_15_corr)))
            axes[0].set_yticklabels([f[:30] for f in top_15_corr.index], fontsize=8)
            axes[0].set_xlabel('Corr√©lation absolue avec Value', fontsize=11)
            axes[0].set_title('Top 15 - Toutes donn√©es', fontsize=12, fontweight='bold')
            axes[0].grid(axis='x', alpha=0.3)
            axes[0].invert_yaxis()
            
            # Mortality Rate < 15
            axes[1].barh(range(len(top_15_corr_filtered)), top_15_corr_filtered.values, color='coral', edgecolor='black', alpha=0.7)
            axes[1].set_yticks(range(len(top_15_corr_filtered)))
            axes[1].set_yticklabels([f[:30] for f in top_15_corr_filtered.index], fontsize=8)
            axes[1].set_xlabel('Corr√©lation absolue avec Value', fontsize=11)
            axes[1].set_title('Top 15 - Mortality < 15', fontsize=12, fontweight='bold')
            axes[1].grid(axis='x', alpha=0.3)
            axes[1].invert_yaxis()
            
            # Mortality < 15 ET GDP < 0.5
            axes[2].barh(range(len(top_15_corr_gdp)), top_15_corr_gdp.values, color='green', edgecolor='black', alpha=0.7)
            axes[2].set_yticks(range(len(top_15_corr_gdp)))
            axes[2].set_yticklabels([f[:30] for f in top_15_corr_gdp.index], fontsize=8)
            axes[2].set_xlabel('Corr√©lation absolue avec Value', fontsize=11)
            axes[2].set_title('Top 15 - Mortality<15 & GDP<0.5', fontsize=12, fontweight='bold')
            axes[2].grid(axis='x', alpha=0.3)
            axes[2].invert_yaxis()
            
            plt.tight_layout()
            plt.show()
            
            # Matrice de corr√©lation pour top features (donn√©es doublement filtr√©es)
            selected_features_gdp = list(top_15_corr_gdp.head(10).index) + ['Value']
            corr_matrix_gdp = df_mortality_gdp_filtered[selected_features_gdp].corr()
            
            # Heatmap
            plt.figure(figsize=(12, 10))
            sns.heatmap(corr_matrix_gdp, annot=True, fmt='.2f', cmap='coolwarm', center=0,
                        square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
                        xticklabels=[f[:28] for f in corr_matrix_gdp.columns],
                        yticklabels=[f[:28] for f in corr_matrix_gdp.index])
            plt.title('Matrice de corr√©lation - Mortality < 15 & GDP < 0.5', fontsize=14, fontweight='bold')
            plt.xticks(rotation=45, ha='right', fontsize=9)
            plt.yticks(fontsize=9)
            plt.tight_layout()
            plt.show()
        else:
            print("‚ö†Ô∏è Colonne 'Value' non trouv√©e")
    else:
        print("‚ö†Ô∏è Aucune donn√©e apr√®s double filtrage!")
else:
    print(f"‚ö†Ô∏è Colonnes manquantes:")
    if mortality_col not in df.columns:
        print(f"   - {mortality_col}")
    if gdp_col not in df.columns:
        print(f"   - {gdp_col}")
        print(f"\nColonnes contenant 'GDP':")
        gdp_cols = [col for col in df.columns if 'gdp' in col.lower()]
        for col in gdp_cols:
            print(f"  - {col}")
```
